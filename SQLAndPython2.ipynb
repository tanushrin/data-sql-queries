```python
import sqlite3

conn = sqlite3.connect('data/movies.sqlite')
db = conn.cursor()
```

```python
query = "SELECT m.title, m.genres, d.name from movies m, directors d"
db.execute(query)
result = db.fetchone()
print(result[0])
```

```python
query = "SELECT m.title, m.genres, d.name \
    from movies m \
    JOIN directors d ON m.director_id = d.id "
db.execute(query)
results = db.fetchall()
print(results)

#print([each_director[0] for each_director in results])
```

```python
query = "SELECT m.title from movies m \
 JOIN directors d ON m.director_id = d.id \
 WHERE  m.start_year > d.death_year \
 ORDER BY m.title"
db.execute(query)
results = db.fetchall()
#print(results)

print([movie[0] for movie in results])
```

```python
gn = 'Action,Adventure,Comedy'

query = f"SELECT COUNT(m.id), AVG(m.minutes) FROM movies m WHERE m.genres LIKE '{gn}'"
db.execute(query)
results = db.fetchall()
print(results)
dict1={}
dict1['genre'] = gn
dict1['number_of_movies'] = results[0][0]
dict1['avg_length'] = round(results[0][1],3)
print(dict1)
```

```python
genre = 'Action,Adventure,Comedy'
```

```python
query = f"SELECT d.name  , COUNT(m.title) as no_of_movies \
 FROM movies m \
 JOIN directors d  ON m.director_id = d.id \
 WHERE m.genres = '{genre}' \
 GROUP BY d.name \
 ORDER BY no_of_movies DESC , d.name ASC \
 LIMIT 5"
db.execute(query)
results = db.fetchall()
print(results)
```

```python
query = "SELECT d.name, \
(m.start_year-d.birth_year) as age_when_first_time_director \
FROM movies m \
JOIN directors d   ON m.director_id = d.id \
WHERE d.birth_year IS NOT NULL \
ORDER BY age_when_first_time_director ASC LIMIT 5"
db.execute(query)
results = db.fetchall()
print(results)
```

```python
query = "SELECT MAX(m.minutes) FROM movies m"
db.execute(query)
results = db.fetchone()
print(results[0])
```
